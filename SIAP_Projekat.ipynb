```python
mask = df_newer_games.notnull()
  
# using dropna() method to drop NaN 
# value rows
tmp = df_newer_games.where(mask).dropna()
tmp.loc[0]
tmp.info()
```

```python
tmp.drop(["Name"], axis = 1, inplace = True)
tmp["User_Score"] = tmp["User_Score"].astype(float)
```

```python
tmp.info()
```

```python
tmp['Genre'] = tmp['Genre'].astype('category').cat.codes
tmp['Publisher'] = tmp['Publisher'].astype('category').cat.codes
tmp['Developer'] = tmp['Developer'].astype('category').cat.codes
tmp['Rating'] = tmp['Rating'].astype('category').cat.codes
import numpy as np
tmp['User_Score'] = tmp['User_Score'].apply(np.ceil)

tmp.tail()
```

```python
tmp['Platform'] = tmp['Platform'].astype('category').cat.codes
```

```python
from sklearn.model_selection import train_test_split

training, test = train_test_split(tmp, test_size = 0.25, random_state = 32)
```

```python
features = training.drop(['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'], axis = 1).columns
print(features)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor

training_features, training_labels = training[features], training['NA_Sales']
test_features, test_labels = test[features], test['NA_Sales']

pipeline = Pipeline([('scaler', MinMaxScaler()), ('regressor', XGBRegressor())])
pipeline.fit(training_features, training_labels)

r2 = pipeline.score(test_features, test_labels)
print(f'Score {r2}')
```

```python
from sklearn import svm
tmp['User_Score'] = tmp['User_Score'].astype(float)
tmp['User_Score'] = np.where(tmp['User_Score'] <=3.5, 0, tmp['User_Score'])
tmp['User_Score'] = np.where((3.5 < tmp['User_Score']) & ( tmp['User_Score'] <=7.5), 1 ,tmp['User_Score'])
tmp['User_Score'] = np.where(7.5 < tmp['User_Score'] , 2 ,tmp['User_Score'])
tmp['User_Score'] = tmp['User_Score'].astype('category').cat.codes
training, test = train_test_split(tmp, test_size = 0.25, random_state = 32)
features = training.drop(['User_Score'], axis = 1).columns

training_features, training_labels = training[features], training['User_Score']
test_features, test_labels = test[features], test['User_Score']

training_features.head()
```

```python
training_labels.head()
```

```python
from sklearn import preprocessing
clf_svm = svm.SVC(kernel='rbf')
training['Platform'] = tmp['Platform'].astype('category').cat.codes
scaler = preprocessing.StandardScaler().fit(training[features])
features_nomalizovane = scaler.transform(training[features])
clf_svm.fit(features_nomalizovane, training_labels)
```

```python
test['Platform'] = tmp['Platform'].astype('category').cat.codes
clf_svm.predict(scaler.transform([test[features].iloc[0]]))
```

```python
clf_svm.score(scaler.transform(test[features]), test_labels)
```

```python
svm_predictions = clf_svm.predict(scaler.transform(test[features]))

from sklearn.metrics import f1_score

f1_score(test_labels, svm_predictions, average=None)
```

```python
from sklearn.metrics import classification_report

print(classification_report(test_labels,svm_predictions))
```
