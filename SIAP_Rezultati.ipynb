```python
df_games_igdb = df_games_igdb.drop_duplicates(subset=["name"], keep='first')
len(df_games_igdb)
```

```python
df_new_combined.drop('hypes', axis=1, inplace = True)
df_new_combined.drop('storyline', axis=1, inplace = True)
df_new_combined.drop('videos', axis=1, inplace = True)
df_new_combined.drop('name', axis=1, inplace = True)
```

```python
mask = df_newer_games.notnull()
  
tmp = df_newer_games.where(mask).dropna()
tmp.loc[0]
tmp.info()
```

```python
tmp.info()
```

```python
tmp.drop(["Name"], axis = 1, inplace = True)
tmp["User_Score"] = tmp["User_Score"].astype(float)

tmp['Platform'] = tmp['Platform'].astype('category').cat.codes
tmp['Genre'] = tmp['Genre'].astype('category').cat.codes
tmp['Publisher'] = tmp['Publisher'].astype('category').cat.codes
tmp['Developer'] = tmp['Developer'].astype('category').cat.codes
tmp['Rating'] = tmp['Rating'].astype('category').cat.codes

tmp['User_Score'] = tmp['User_Score'].apply(np.ceil)
tmp.info()
```

```python
from sklearn.model_selection import train_test_split

training, test = train_test_split(tmp, test_size = 0.25, random_state = 32)

features = training.drop(['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'], axis = 1).columns
print(features)
```
